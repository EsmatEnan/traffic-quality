```python
#importing numpy and osmnx for dependencies (map-matching) 

import numpy as np
import osmnx as ox


lat = 51.950296032982756
lng = 7.615756988525391

# Get the OSM graph of "drive" streets for münster which will be used by leuvan mapmatching and plot it
graph = ox.graph_from_point((lat, lng), dist=5000, network_type='drive')
ox.plot_graph(graph, fig_height=10, fig_width=10)
```

```python
#importing leuvan mapmatching for map-matching

from leuvenmapmatching.matcher.distance import DistanceMatcher
from leuvenmapmatching.map.inmem import InMemMap
from leuvenmapmatching.matcher.newsonkrumm import NewsonKrummMatcher

# Leuven Map Matching is using a different internal graph structure for the street data. 
# Therefore, the OSMnx graph needs to be transformed to the InMemMap
streetmap = InMemMap("enviroCar", use_latlon=True, use_rtree=True, index_edges=True)

# add nodes
nodes = list(graph.nodes)
for node in nodes:
    #if graph.nodes[node]['highway'] == 'traffic_signals':
    lng = graph.nodes[node]['x']
    lat = graph.nodes[node]['y']
    streetmap.add_node(node, (lat, lng))
    #print(graph.nodes[node]['highway'])

# add edges
edges = list(graph.edges)
for edge in edges:
    node_a, node_b = edge[0], edge[1]
    streetmap.add_edge(node_a, node_b)
    #print(graph.edges[edge]['name'])
    # exclude bi-directional edges when street is oneway
    if not graph.edges[edge]['oneway']:
        streetmap.add_edge(node_b, node_a)
    
streetmap.purge()
```

```python
# Configure the mapmatcher. For increasing the performance it is important 
# that max_dist (in meters) is not set too high
matcher = NewsonKrummMatcher(streetmap,
                          max_dist=100, 
                          max_dist_init=200,
                          #non_emitting_length_factor=0.75,
                          obs_noise=100,
                          #obs_noise_ne=75,
                          dist_noise=5,
                          non_emitting_states=False,
                          non_emitting_edgeid=False)
```

```python
#function for map-matching

def mm_coords(track, matcher):
    
    """
    DOC STRING
    
    Parameters:
    track  - Dataframe of a track for which you want to perform map-matching
    matcher- Matcher created using the Newson Krumm Matcher for the osm graph
    
    Usage:
    performs map-matching and adds map-matched coordinates and edge information to the track data
    
    Output:
    Track data with two columns added "MM_coords" and "Edge" for map-matched coordinates and osm edge id respectively
    
    """
    
    # Get the latitude and longitude coordinates from the track and join them in an array
    lats = list(track['geometry'].apply(lambda coord: coord.y))
    lngs = list(track['geometry'].apply(lambda coord: coord.x))

    track_coords = [c for c in zip(lats, lngs)]
    
    edge_ids, last_idx = matcher.match(track_coords)
    coords = [m.edge_m.pi[:] for m in matcher.lattice_best]
    
    i = 0
    edg = []
    for edge in edge_ids:
        index = edge_ids[i]
        tup = list(index)
        tup.append(0)
        ed = tuple(tup)
        edg.append(ed)
        i = i+1
        
    track['MM_coords'] = coords
    track['Edge'] = edg
```

```python
#function for parsing the remapped shapefile and making it into a dataframe

def read_shapefile(shp_path):
    
    """
    DOC STRING
    
    Parameters:
    shp_path- path location to the shapefile which needs to be parsed
    
    Usage:
    parses the shapefile and makes it into a dataframe
    
    Output:
    Pandas Dataframe which was created using the shapefile
    
    """

    import shapefile

    #read file, parse out the records and shapes
    sf = shapefile.Reader(shp_path)
    fields = [x[0] for x in sf.fields][1:]
    records = sf.records()
    shps = [s.points for s in sf.shapes()]
    #write into a dataframe
    df = pd.DataFrame(columns=fields, data=records)
    df = df.assign(coords=shps)
    return df
```

```python
#FUNCTIONS FOR ANALYSIS


#cars waiting(stopped car) in the segment
def no_of_waiting_cars(edge):
    
    """
    DOC STRING
    
    Parameters:
    edge- ID of the road segment
    
    Usage:
    returns the no of cars that stopped (speed < 5km) while passing through this segment
    
    Output:
    Total no of cars which has stopped atleast once while travelling through a segment (Integer)
    
    """
    
    uniq = tracks[(tracks['id_2'] == edge)]   #id_2 is the edge id of our defined network 
    uniq1= uniq[(uniq['Speed.valu'] < 5)]['track.id'].unique()
    num1 = len(uniq1)
    return num1 
    
#average speed of tracks
def avg_speed(edge):
    
    """
    DOC STRING
    
    Parameters:
    edge- ID of the road segment
    
    Usage:
    returns the average speed of the cars which has passed through this segment
    
    Output:
    Average speed of the cars which has passed through this segment (Float)
    
    """
    speed = tracks[(tracks['id_2'] == edge)]['Speed.valu'].mean()  #id_2 is the edge id of our defined network
    return speed

#no.of.stops
def no_of_stops(edge, trackid):
    
    """
    DOC STRING
    
    Parameters:
    edge- ID of the road segment
    track_id- ID of the track (not ID of a particular point in the track) 
    
    Usage:
    returns the number of stops made by a single track while travelling through this segment
    
    Output:
    Number of stops made by a single track while travelling through this segment (Integer)
    
    Note:
    In this function, it is assumed that a stop a constitutes when the speed of a car is less than 5km/hr
    If the speed is less than 5 km/hr consecutively for a single track, then all the consecutive speed 
    which is less than 5 km/hr is considered as a single stop instead of a number of stops
    
    """
    
    
    count = 0
    track_id= tracks[tracks['track.id']==trackid]
    track_seg= track_id[(track_id['id_2'] == edge)]  #id_2 is the edge id of our defined network
    list_of_speedS = list(track_seg['Speed.valu'])
    for i in range(len(list_of_speedS)):
        if i!=0 and list_of_speedS[i]<5:
            if list_of_speedS[i-1]<5:
                continue
            else:
                count=count+1              
        elif i==0 and list_of_speedS[i]<5:
            count=count+1
    return count

#no.of.cars.passing.the.segment
def no_of_cars_passing(edge):
    
    
    """
    DOC STRING
    
    Parameters:
    edge- ID of the road segment
    
    Usage:
    returns the number of cars that has traversed through this segment
    
    Output:
    Total number of cars that has traversed through this segment (Integer)
    
    """
    
    no_of_cars = tracks[(tracks['id_2'] == edge)]['track.id'].unique()  #id_2 is the edge id of our defined network
    no = len(no_of_cars)
    return no

#maxspeed exceedance
def speed_limit(edge):
    
    """
    DOC STRING
    
    Parameters:
    edge- ID of the road segment
    
    Usage:
    returns the speed limit for the given segment
    
    Output:
    Speed limit for the given segment (Integer)
    
    """
    
    max_speed = tracks[(tracks['id_2'] == edge)]['maxspeed'].unique()  #id_2 is the edge id of our defined network
    max_speed= float(max_speed[0])     #maxspeed is the speed limit of the road segment
    return max_speed
```

```python
#no.of.individual exceeded cases per segment
def individual_cases(edge, track_id):
    
    
    """
    DOC STRING
    
    Parameters:
    edge- ID of the road segment
    track_id- ID of a particular track
    
    Usage:
    returns the number of cases where the speed limit has been exceeded by cars for the given segment
    
    Output:
    Number of cases where the speed limit has been exceeded by cars for the given segment (Integer)
    
    Note:
    In this function, when the speed of a car is above the maxspeed (speed limit) continuously 
    for a single track  then all the consecutive speed which is above the speed limit is considered
    as a single case instead of a number of cases
    
    """
    
    count = 0
    edge_df= tracks[(tracks['id_2'] == edge)]  #id_2 is the edge id of our defined network
    max_speed = speed_limit(edge)
    edge_track= edge_df[edge_df['track.id']==track_id]
    list_of_speedS = list(edge_track['Speed.valu'])
            
    for i in range(len(list_of_speedS)):
        if i!=0 and list_of_speedS[i]> max_speed:
                    
            if list_of_speedS[i-1]> max_speed:
                continue
            else:
                count=count+1        
        elif i==0 and list_of_speedS[i]> max_speed:
                    count=count+1
            
    return count
```

```python
#no.of.individual exceeded cases per segment
def individual_cases_id(edge, track_id):
    
    """
    DOC STRING
    
    Parameters:
    edge- ID of the road segment
    track_id- ID of a particular track
    
    Usage:
    returns list of point id in tracks where the speed limit has been exceeded for the given segment (Warning: contains empty lists where
    speed limit isn't exceeded)
    
    Output:
    Number of cases where the speed limit has been exceeded by cars for the given segment (Integer)
    
    Note:
    In this function, when the speed of a car is above the maxspeed (speed limit) continuously 
    for a single track  then the first time the speed reaches above the speed limit is considered
    instead of a number of times
    
    """
    
    point_id = []
    edge_df= tracks[(tracks['id_2'] == edge)]   #id_2 is the edge id of our defined network
    edge_track_df= edge_df[edge_df['track.id']==track_id]  
    max_speed = speed_limit(edge)
    edge_track= list(edge_track_df['id'])
    list_of_speedS = list(edge_track_df['Speed.valu'])
            
    for i in range(len(list_of_speedS)):
        if i!=0 and list_of_speedS[i]> max_speed:
                    
            if list_of_speedS[i-1]> max_speed:
                continue
            else:
                point_id.append(edge_track[i])
                        
        elif i==0 and list_of_speedS[i]> max_speed:
            point_id.append(edge_track[i])
            
    return point_id
```

```python
#performing individual_cases_id
id_indi_excedCases=[]
for i in segment_numbers:
    cases = []
    for j in track_numbers:
        a=individual_cases_id(i,j)
        cases.append(a)
    id_indi_excedCases.append(cases)
```

```python
#function for finding the total duration of stops in a segment for a track

def waiting_time(edge, trackid):
    
    """
    DOC STRING
    
    Parameters:
    edge- ID of the road segment
    track_id- ID of a particular track
    
    Usage:
    returns the total duration of stops made by a single track on a given segment
    
    Output:
    Total duration of stops made by a single track on a given segment (Integer)
    
    Note:
    In this function, when the speed of a car is above the maxspeed (speed limit) continuously 
    for a single track  then the first time the speed reaches above the speed limit is considered
    instead of a number of times
    
    """
    
    
    w_time=0
    track_id= tracks[tracks['track.id']==trackid]
    track_seg= track_id[(track_id['id_2'] == edge)]  #id_2 is the edge id of our defined network
    list_of_speedS = list(track_seg['Speed.valu'])
    FMT = '%H:%M:%S'
    track_time=[]
    
    for i in track_seg['new_time']:
        a = datetime.strptime(i, FMT)
        track_time.append(a)
        
        
        
    for i in range(len(list_of_speedS)):
        FMT = '%H:%M:%S'
        if i!=(len(list_of_speedS)-1) and list_of_speedS[i]<5:
            if list_of_speedS[i+1]>=5:
                tdelta= track_time[i+1]-track_time[i]
                a=tdelta.total_seconds()
                w_time= w_time+abs(a)
            elif list_of_speedS[i+1]<5:
                continue
                
        elif i==(len(list_of_speedS)-1):
            continue
            
    return w_time
```

```python
#creating lists for total duration of stops for each track in all segments

stop_duration = []
for i in segment_numbers:
    ave_segment_time=[]
    time = 0
    for j in track_numbers:
        a= waiting_time(i,j)
        ave_segment_time.append(a)
        
    for k in ave_segment_time:
            time= time + k
        
    stop_duration.append(time/60)
```

```python
Segment_df['Total_stop_duration (mins)']=stop_duration
```
